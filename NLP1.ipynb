```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/nlp_preprocessing_dataset.csv")
```

```python
df
```

```python
df["text"]=df["text"].str.lower()
```

```python
df
```

```python
import re
```

```python
def remove_html_tags(text):
  pattern="<.*?>"
  clean_text=re.sub(pattern, "", text)
  return clean_text
```

```python
df["new_text"]=df["text"].apply(remove_html_tags)
```

```python
df
```

```python
def removal_url(text):
    pattern="https?://\S+|www\.\S+"
    clean_text=re.sub(pattern,"",text)
    return clean_text
```

```python
df["new_text"]=df["new_text"].apply(removal_url)
```

```python
df
```

```python
def remove_newline(text):
    pattern="\n"
    clean_text=re.sub(pattern,"",text)
    return clean_text
```

```python
df["new_text"]=df["new_text"].apply(remove_newline)
```

```python
df
```

```python
import string
```

```python
exclude=list(string.punctuation)
```

```python
exclude
```

```python
def remove_punctuation(text):
    # This pattern matches one or more occurrences of any character in string.punctuation
    # re.escape is used to escape any special regex characters within string.punctuation
    punctuation_pattern = r'[' + re.escape(string.punctuation) + r']+'
    clean_text = re.sub(punctuation_pattern, "", text)
    return clean_text
```

```python
df["new_text"]=df["new_text"].apply(remove_punctuation)
```
