```python
def one_hot_encoding(text):
    words=text.lower().split()
    vocab=list(set(words))
    word_to_index={}
    for i in range(len(vocab)):
        word_to_index[vocab[i]]=i
    result=[]
    for x in words:
        vector=[0]*len(vocab)
        index=word_to_index[x]
        vector[index]=1
        result.append(vector)
    return result
```

```python
text="pasta is good and delecious"
one_hot_encoding(text)
```

```python
def bag_of_words(text):
    words = text.lower().split()
    vocab = list(set(words))
    d = {}
    for i in range(len(vocab)):
        d[vocab[i]] = i
    vector = [0] * len(words)
    cnt = {}
    for i in words:
        if i not in cnt:
            cnt[i] = 1
        else:
            cnt[i] +=1
    return vector
```

```python
text="pasta is good and delecious"
bag_of_words(text)
```
