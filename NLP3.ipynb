```python
import re
def removal_html_tags(text):
    pattern="<.*?>"
    clean_text=re.sub(pattern,"",text)
    return clean_text
```

```python
def removal_url(text):
    pattern="https?://\S+|www\.\S+"
    clean_text=re.sub(pattern,"",text)
    return clean_text
```

```python
df["email_text"]=df["email_text"].apply(removal_url)
```

```python
import string
```

```python
exlcude=list(string.punctuation)
```

```python
exlcude
```

```python
def remove_punctuations(text):
    for char in exlcude:
        if char in text:
            text=text.replace(char,"")
    return text
```

```python
df["email_text"]=df["email_text"].apply(remove_punctuations)
```

```python
df.head()
```

```python
def remove_newline_character(text):
    pattern="\n"
    clean_text=re.sub(pattern,"",text)
    return clean_text
```

```python
df["email_text"]=df["email_text"].apply(remove_newline_character)
```

```python
df.head()
```

```python
def remove_extra_space(text):
    pattern="\s+"
    clean_text=re.sub(pattern," ",text)
    return clean_text
```

```python
df["email_text"]=df["email_text"].apply(remove_extra_space)
```

```python
chatwords={
    "u":"you",
    "omg":"oh my god",
    "r":"are",
    "plzzz":"please",
    "ur":"you are",
    "ill":"I will"
}
```

```python
def correcting_chatwords(text):
    words=text.split()
    new_words=[]
    for word in words:
        if word in chatwords:
            new_words.append(chatwords[word])
        else:
            new_words.append(word)
    return " ".join(new_words)

```

```python
txt="omg how r u"
correcting_chatwords(txt)
```

```python
df["email_text"]=df["email_text"].apply(correcting_chatwords)
```

```python
df1=pd.read_csv("/content/urdu_noisy_sentiment_dataset.csv")
```

```python
df1
```

```python
text_col = "text"      # Urdu sentence
label_col = "sentiment"   # sentiment (positive/negative)
```

```python
df1[label_col] = df1[label_col].str.lower()

# Example mapping
df1[label_col] = df1[label_col].map({
    "positive": 1,
    "negative": 0
})
```

```python
print(df1.head())
print(df1.isnull().sum())
print(df1[label_col].value_counts())
```

```python
def clean_urdu_text(text):
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove English words (optional)
    text = re.sub(r'[a-zA-Z]', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s\u0600-\u06FF]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df1[text_col] = df1[text_col].apply(clean_urdu_text)
```

```python
df1.head()
```

```python
df1
```

```python
urdu_stopwords = [
    "ہے", "ہیں", "کو", "میں", "کا", "کی", "کے", "اور",
    "سے", "پر", "یہ", "وہ", "کہ", "کر", "ہی"
]

def remove_stopwords(text):
    words = text.split()
    words = [w for w in words if w not in urdu_stopwords]
    return " ".join(words)
```

```python
df1[text_col] = df1[text_col].apply(remove_stopwords)
```

```python
df1[label_col] = df1[label_col].str.lower()

# Example mapping
df1[label_col] = df1[label_col].map({
    "positive": 1,
    "negative": 0
})
```

```python
df1 = df1.drop_duplicates(subset=[text_col])
```

```python
print(df1.head())
print(df1.isnull().sum())
print(df1[label_col].value_counts())
```

```python
df1.to_csv("cleaned_urdu_sentiment.csv", index=False)
```

```python
def normalize_urdu(text):
    text = re.sub(r'ي', 'ی', text)
    text = re.sub(r'ك', 'ک', text)
    text = re.sub(r'ة', 'ہ', text)
    return text

df1[text_col] = df1[text_col].apply(normalize_urdu)
```

```python
df2=pd.read_csv("/content/marathi_noisy_sentiment.csv")
```

```python
df2
```

```python
text_col = "text"
label_col = "label"
```

```python
df2 = df2.dropna(subset=[text_col])
df2[text_col] = df2[text_col].astype(str)
```

```python
def clean_marathi_text(text):
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove English characters (optional)
    text = re.sub(r'[a-zA-Z]', '', text)

    # Keep only Marathi (Devanagari Unicode range)
    text = re.sub(r'[^\u0900-\u097F\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df2
```

```python
df2[text_col] = df2[text_col].apply(clean_marathi_text)
```

```python
marathi_stopwords = [
    "आहे", "आहेत", "मी", "तू", "तो", "ती", "ते",
    "आणि", "किंवा", "पण", "म्हणून", "वर", "मध्ये",
    "का", "हो", "नाही", "होते", "होता"
]

def remove_stopwords(text):
    words = text.split()
    words = [w for w in words if w not in marathi_stopwords]
    return " ".join(words)

df2[text_col] = df2[text_col].apply(remove_stopwords)
```

```python
def normalize_marathi(text):
    text = re.sub(r'क़', 'क', text)
    text = re.sub(r'ख़', 'ख', text)
    text = re.sub(r'ग़', 'ग', text)
    text = re.sub(r'ज़', 'ज', text)
    text = re.sub(r'फ़', 'फ', text)
    return text

df2[text_col] = df2[text_col].apply(normalize_marathi)
```

```python
label_col = "sentiment" # Update label_col for df2
df2[label_col] = df2[label_col].str.lower()

df2[label_col] = df2[label_col].map({
    "positive": 1,
    "negative": 0,
    "neutral": 2   # if exists
})
```

```python
df2 = df2.drop_duplicates(subset=[text_col])
```

```python
print(df2.head())
print(df2.isnull().sum())
print(df2[label_col].value_counts())
```

```python
df2.to_csv("cleaned_marathi_sentiment.csv", index=False)
```
