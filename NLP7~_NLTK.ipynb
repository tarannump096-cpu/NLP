```python
from nltk.corpus import stopwords
```

```python
stop_words=list(stopwords.words("english"))
```

```python
stop_words
```

```python
def stopwords_removal(sentence):
    text=sentence.lower()
    words=word_tokenize(text)
    res=[]
    for word in words:
        if word not in stop_words:
            res.append(word)
    return " ".join(res)
```

```python
sen11="this is a python class"
```

```python
stopwords_removal(sen11)
```

```python
from nltk.stem import PorterStemmer

# Initialize the stemmer
ps = PorterStemmer()

# Example: Stemming individual words
words = ["program", "programs", "programmer", "programming", "programmers"]
for w in words:
    print(w, ":", ps.stem(w))
```

```python
def stemming(text):
    text=text.lower()
    tokenized=word_tokenize(text)
    res=[]
    for word in tokenized:
        stem_word= ps.stem(word)
        res.append(stem_word)
    return res
```

```python
text="walking playing swimming"
stemming(text)
```
